```python
import scipy as sp
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1) x(t) = sin 2pi 3t + sin 2pi 7pi t + sin 2pi 11t
Fs = 1000
t = np.arange(0,2,1./Fs)
x = np.sin(2*np.pi*3*t) + np.sin(2*np.pi*7*t) + np.sin(2*np.pi * 11 * t)
```

```python
X = sp.fft(x)
```

```python
f = np.linspace(0,Fs,len(x))
```

```python
plt.stem(f[0:50],np.abs(X[0:50]))
```
