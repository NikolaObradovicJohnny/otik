```python
import scipy as sp
from scipy.io.wavfile import read
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
(fs,x)=read('raccoon1.wav')
```

```python
len(x)
```

```python
plt.plot(x)
```

```python
X=sp.fft(x)
```

```python
f=np.linspace(0,fs,len(x))
```

```python
plt.stem(f[0:1000],np.abs(X[0:1000]))
```

```python
np.shape(x)
```

```python
from scipy import signal
b,a=signal.butter(5,250/(fs/2),btype='highpass')
```

```python
filtered_signal=signal.lfilter(b,a,x)
```

```python
write('filteredsignal.wav',fs,filtered_signal) #possible to open with aimp3 software
```

```python
noise=np.random.randn(len(x))
```

```python
signal_with_noise=x+noise
```

```python
plt.plot(signal_with_noise)
```

```python
write('signalwithnoise.wav',fs,signal_with_noise)
```

```python
plt.plot(noise)
```

```python
write('noise.wav',fs,noise)
```

```python
b,a=signal.butter(5,250/(fs/2),btype='lowpass')
```

```python
filtered_signal_with_noise=signal.lfilter(b,a,signal_with_noise)
```

```python
write('filtered_signal_with_noise.wav',fs,filtered_signal_with_noise)
```
